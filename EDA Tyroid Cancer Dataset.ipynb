```python
#Encode Categorical Variables (for correlation)
df_encoded = pd.get_dummies(df, drop_first=True)

```

```python
#Age Distribution
plt.hist(df["Age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

Most patients fall in middle to higher age groups → age is critical.

```python
#Gender Distribution
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

The dataset has unequal gender representation, which may influence cancer risk analysis.

```python
#Age Outlier Detection
sns.boxplot(y="Age", data=df)
plt.title("Age Outlier Detection")
plt.show()
```

Older age outliers exist and represent valid high-risk medical cases rather than errors.

```python
#Age vs Cancer Risk
sns.boxplot(x=df.iloc[:, -1], y="Age", data=df)
plt.title("Age vs Cancer Risk")
plt.show()
```

Cancer-positive patients generally have a higher median age than non-cancer patients.

```python
#Gender vs Cancer Risk
sns.countplot(x="Gender", hue=df.iloc[:, -1], data=df)
plt.title("Gender vs Cancer Risk")
plt.show()
```

Cancer occurrence varies by gender, indicating gender as a meaningful predictive feature.

```python
#Age Density by Cancer Risk (KDE Plot)
sns.kdeplot(data=df, x="Age", hue=df.columns[-1], fill=True)
plt.title("Age Density by Cancer Risk")
plt.show()
```

Cancer cases are concentrated at higher ages, though some overlap exists with non-cancer cases.

```python
#Scatter Plot – Age vs Numeric Feature
num_col = df.select_dtypes(include=np.number).columns[1]
sns.scatterplot(x="Age", y=num_col, hue=df.columns[-1], data=df)
plt.title("Scatter Plot: Age vs Feature")
plt.show()
```

Cancer cases form visible clusters, showing interaction between age and medical indicators.

```python
#Correlation Heatmap
df_encoded = pd.get_dummies(df, drop_first=True)
sns.heatmap(df_encoded.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Several features show meaningful correlation with thyroid cancer diagnosis.

STRONG INSIGHTS 

Age alone increases risk, but age + abnormal indicators increase it sharply

Certain categorical groups consistently show higher cancer proportions

Numerical indicators shift upward for cancer patients

Feature interactions matter more than isolated values

Dataset reflects realistic medical complexity
